```python
import pandas as pd
import numpy as np
```

```python
filepath = '/Users/maiya/Downloads/'
filename = 'rock_paper_scissors.csv'

rock_df = pd.read_csv(filepath+filename)
```

```python
rock_df.head()
```

```python
rock_df = (
rock_df
.assign(
    opponent = lambda x: np.select(
        [(x.Opponent=='A'), (x.Opponent=='B'), (x.Opponent=='C'), True],
        ['rock', 'paper', 'scissors', '?'])
).assign(
        me = lambda x:np.select(
            [(x.Me=='X'), (x.Me=='Y'), (x.Me=='Z'), True],
            ['rock', 'paper', 'scissors', '?'])
    )
).drop(
columns=['Opponent', 'Me', 'score'])

#company = lambda x: np.where(x['company']=='vw', 'volk', x['company'])
#lambda x:np.select(
#             [(x['cylinders']==3), (x['cylinders']==4) & (x['mpg']>=30),(x['cylinders']==5),(x['cylinders']==4) & (x['mpg']<30),(x['cylinders']==6),(x['cylinders']==8),True],
#             ['Hatchback','Hatchback','Sedan','Sedan','SUV','Sports','?']
```

```python
rock_df = (
rock_df
    .assign(
    score = lambda x: np.where(
    x['opponent']==x['me'], 3, 'NaN'))
)
 
```

```python
rock_df.head()
```

```python
rock_df.sample(10)
```

```python
rock_df = (
rock_df
    .assign(
        score = lambda x: np.where(
        (x['opponent']=='rock') & (x['me']=='scissors'),
        0, x['score']
        )
    )
    .assign(
        score = lambda x: np.where(
        (x.opponent == 'rock') & (x.me == 'paper'),
            6, x.score)
    )
    .assign(
        score = lambda x: np.where(
        (x.opponent=='paper') & (x.me == 'scissors'),
            6, x.score)
    )
    .assign(
        score = lambda x: np.where(
        (x.opponent=='paper')&(x.me=='rock'),
        0, x.score)
    )
    .assign(
        score = lambda x: np.where(
        (x.opponent=='scissors')&(x.me=='paper'),
            0, x.score
        )
    )
    .assign(
        score = lambda x: np.where(
        (x.opponent=='scissors')&(x.me=='rock'),
            6, x.score
        )
    )
)
```

```python
rock_df.sample(10)
```

```python
rock_df = (
rock_df
    .assign(
    add_score = lambda x: np.select(
        [x.me=='rock', x.me=='paper', x.me=='scissors'],
        [1,2,3]
    )
    )
)


#lambda x:np.select(
#             [(x['cylinders']==3), (x['cylinders']==4) & (x['mpg']>=30),(x['cylinders']==5),(x['cylinders']==4) & (x['mpg']<30),(x['cylinders']==6),(x['cylinders']==8),True],
#             ['Hatchback','Hatchback','Sedan','Sedan','SUV','Sports','?']
```

```python
rock_df.score = rock_df.score.astype(int)
```

```python
rock_df['final_score'] = rock_df['score'] + rock_df['add_score']
```
